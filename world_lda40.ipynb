```python
import warnings
warnings.filterwarnings("ignore")

from gensim.models import ldamodel
from gensim import corpora

import pyLDAvis
from pyLDAvis import gensim as ldavis
pyLDAvis.enable_notebook()
```

```python
fp = "D:\\Projects\\theGuardian\\"

dict_path = fp + "dictionary_world_words.dict"
dictionary = corpora.Dictionary.load(dict_path)

corpus_path = fp + "world_articles_corpus.mm"
corpus = corpora.MmCorpus(corpus_path)

lda_path = fp + "topics40world"
lda = ldamodel.LdaModel.load(lda_path, mmap='r')
```

```python
vis_data = ldavis.prepare(lda, corpus, dictionary)
pyLDAvis.display(vis_data)
```
